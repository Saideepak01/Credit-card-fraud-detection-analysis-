# Seperating legit and fraud transactions

# Under-Sampling the dataset

```python
#Since the dataset is highly unbalanced we will be under-sampling the dataset which a takes a sample of the data. So, we will be generating a sample dataset from the original dataset containing the same distribtion of legit and fraud transactions
```

```python
legit_sample = legit.sample(n=492) #Of the 284315 legit transactions we will the randomely taking 492 transactions in this sample and combine it with fraud transations
```

# Contatenating two dataframes

```python
lf_data = pd.concat([legit_sample, fraud], axis=0)
```

# Splitting dataset into features(time, amount, v1, v2 etc...) and target(class)

```python
x = lf_data.drop(columns='Class', axis=1)
y = lf_data['Class']
```

```python
x
```

```python
y
```

# Importing train_test_split library

```python
from sklearn.model_selection import train_test_split
```

# Splitting train and test data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
# We have split the dataset randomely using random_state into training and testing dataset for the test size we have seperated 80% into training data and remaining 20% into test data the stratify is for evenly distributing the y dataset into train and test 
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

```python
print(y.shape, y_train.shape, y_test.shape)
```

```python
# By viewing the shape we see that about 984 values are in our original x dataset out of this 787 values is given to train and 197 values to test 
```

# Importing the algorithm

```python
# Here we are using logistic regression as we can get a better accuracy score we generally use logistic regression for binary classification types of problems this gives a high degree of accuracy 
```

```python
from sklearn.linear_model import LogisticRegression
```

# fitting and training the data

```python
lrmodel = LogisticRegression()
```

```python
lrmodel.fit(x_train, y_train)
```

# Importing accuracy score library

```python
from sklearn.metrics import accuracy_score #This accuracy score allows us to find wheather the model is correctly predicting or not according to the precentage
```

# Evaluating the trained model

```python
xtprediction = lrmodel.predict(x_train)
tdataaccuracy = accuracy_score(xtprediction, y_train) # we are checking the accuracy of the trained model
```

```python
tdataaccuracy
```

```python
#As the model has been trained from the dataset we have predicted the accuracy for x_train and by comparing it with y_train and we get the accuracy score of 91%
```

# Evaluating the test model

```python
xtestpred = lrmodel.predict(x_test)
testdaccuracy = accuracy_score(xtestpred, y_test)
```

```python
testdaccuracy
```

```python
# The accuracy score of the test data comparing it with x_test and y_test is 94%
```
